```python
# Dependencies and Setup
import pandas as pd
import glob
```

```python
# Read in all 2019 CSVs files of Citi Bike Data into one Dataframe

files = glob.glob("citibike_data\JC-2019*.csv")
dfs = [pd.read_csv(f, header=0, sep=",") for f in files]

bikedata2019 = pd.concat(dfs,ignore_index=True)

# Review file created
bikedata2019.head()
```

```python
# Read in all 2020 CSVs files of Citi Bike Data into one Dataframe

files = glob.glob("citibike_data\JC-2020*.csv")
dfs = [pd.read_csv(f, header=0, sep=",") for f in files]

bikedata2020 = pd.concat(dfs,ignore_index=True)

# Review file created
bikedata2020.head()
```

```python
# Read in all 2021 CSVs files of Citi Bike Data into one Dataframe

files = glob.glob("citibike_data\JC-2021*.csv")
dfs = [pd.read_csv(f, header=0, sep=",") for f in files]

bikedata2021 = pd.concat(dfs,ignore_index=True)

# Review file created
bikedata2021.head()
```

```python
# Combine the dfs with same layout

bikedata1920 = bikedata2019.append(bikedata2020, ignore_index=True)
bikedata1920.head()
```

```python
print(bikedata1920.columns)
print(bikedata2021.columns)
```

```python
# Remove columns which aren't available across all the years and reorder the columns to match across different types

bikedata1920 = bikedata1920[['starttime', 'stoptime', 'start station name','start station id',
       'end station name', 'end station id', 'start station latitude','start station longitude',
       'end station latitude', 'end station longitude']]

#Rename columns to align across data sets
bikedata1920 = bikedata1920.rename(columns={'starttime':'started_at', 'stoptime':'ended_at', 'start station name':'start_station_name','start station id':'start_station_id',
       'end station name': 'end_station_name', 'end station id':'end_station_id','start station latitude':'start_lat','start station longitude':'start_lng',
       'end station latitude':'end_lat', 'end station longitude':'end_lng'})

bikedata2021 = bikedata2021[['started_at', 'ended_at',
       'start_station_name', 'start_station_id', 'end_station_name',
       'end_station_id', 'start_lat', 'start_lng', 'end_lat', 'end_lng']]

# Merge the data together
all_bikedata = bikedata1920.append(bikedata2021, ignore_index=True)
all_bikedata.head()
```

```python
# Create separate data frame with all bike stations

startstation = all_bikedata[[ 'start_station_name', 'start_station_id', 'start_lat', 'start_lng']]
startstation = startstation.rename(columns={'start_station_name':'station_name', 'start_station_id':'station_id', 'start_lat':'lat', 'start_lng':'lng'})
endstation = all_bikedata[[ 'end_station_name','end_station_id', 'end_lat', 'end_lng']]
endstation = endstation.rename(columns={'end_station_name':'station_name', 'end_station_id':'station_id', 'end_lat':'lat', 'end_lng':'lng'})
allstations = startstation.append(endstation, ignore_index=True)
allstations.head()
```

```python
# Remove all duplicates of the stations (excluding the ID as the ID numbers change in 2021)

allstations.drop_duplicates(subset=['station_name', 'lat', 'lng'], keep='first', inplace=True)
```

```python
# Confirm output file and export data into a csv

all_bikedata.to_csv('output_data/all_bikedata.csv',header=True,index=False)
allstations.to_csv('output_data/allstations.csv',header=True,index=False)
```
